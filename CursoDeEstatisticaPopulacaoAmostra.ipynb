Classifica√ßao:Naive Bayes

```python
risco=pd.read_csv('/content/drive/MyDrive/DadosML/credit_data.csv')
#risco.replace({'0_15':'<15.000','15_35':'>=15.000 a <=35.000','acima_35':'>35.000'},inplace=True)
np.unique(risco.iloc[:,-1],return_counts=True)
risco=risco.dropna()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(risco.iloc[:,1:-1], risco.iloc[:,-1] , test_size=0.2, random_state=32)
clf=GaussianNB()
clf.fit(X_train,y_train)
pred=clf.predict(X_test)
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
accuracy_score(y_test,pred)
print(classification_report(y_test,pred))
```

Subamostragem e Sobreamostragem

Undersampling - TomekLinks

```python
from imblearn.under_sampling import TomekLinks
```

```python
tl=TomekLinks(sampling_strategy='majority')#remove a maioria
X_res,Y_res=tl.fit_resample(risco.iloc[:,1:-1],risco.iloc[:,-1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, Y_res , test_size=0.2, random_state=32)
clf2=GaussianNB()
clf2.fit(X_train,y_train)
pred=clf2.predict(X_test)
from sklearn.metrics import accuracy_score
print(classification_report(y_test,pred))
```

Oversampling -SMOTE

```python
from imblearn.over_sampling import SMOTE
```

```python
smt=SMOTE(sampling_strategy='minority')#aumenta a minoria
X_res, Y_res=smt.fit_resample(risco.iloc[:,1:-1],risco.iloc[:,-1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_res, Y_res , test_size=0.2, random_state=32)
clf3=GaussianNB()
clf3.fit(X_train,y_train)
pred=clf3.predict(X_test)
from sklearn.metrics import accuracy_score
print(classification_report(y_test,pred))
```
