```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/EstatisticaDados/census.csv')
data.head()
```

Grafico de dispersao

```python
sns.relplot(data=data,y='final-weight',x='age',hue='income',style='sex',size='education-num')
```

Grafico de barra e setor

```python
sns.barplot(data=data,x='sex',y='final-weight',hue='marital-status')
```

```python
dados_agrupados=data.groupby(['income'])['education-num'].sum()
dados_agrupados.plot.pie()
```

Grafico de linha

```python
sns.boxplot(data=data,x='age')
```

```python
sns.boxplot(data=data,x='education-num')
```

```python
sns.boxplot(data=data,x='hour-per-week')
```

Graficos com atributos categoricos


```python
sns.catplot(data=data.query('age > 70  '),x='sex',y='hour-per-week',hue='income')
```
