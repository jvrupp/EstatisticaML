```python
import pandas as pd
import numpy as np
```

```python
dados = {'emprego': ['Adminstrador_banco_dados', 'Programador', 'Arquiteto_redes'],
         'nova_jersey': [97350, 82080, 112840],
         'florida': [77140, 71540, 62310]}
dados =pd.DataFrame(dados)
```

```python
tot_nj=dados['nova_jersey'].sum()
flor=dados['florida'].sum()
```

```python
dados['nova_jersey%']=dados['nova_jersey'].apply(lambda n : (n*100)/tot_nj)
dados['florida%']=dados['florida'].apply(lambda n : (n*100)/flor)
dados=dados.append({'emprego':'total','nova_jersey':tot_nj,'florida':flor,'nova_jersey%':100,'florida%':100},ignore_index=True)
```

```python
censu = pd.read_csv('/content/drive/MyDrive/EstatisticaDados/census.csv')
censu.head()
```

```python
censuEI=censu.loc[:,['education','income']]
gruposEI=censuEI.groupby(['education']).sum()
gruposEI=gruposEI.drop(index=' Preschool')
```

Exercicio de area com maior chance de ganho

```python
for i in range(gruposEI.shape[0]):
  t1=np.unique(gruposEI.iloc[i,:]['income'].split(' '),return_counts=True)[1][1]
  t2=np.unique(gruposEI.iloc[i,:]['income'].split(' '),return_counts=True)[1][2]
  print((t2*100)/(t1+t2))
```
